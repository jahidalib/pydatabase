```python
#price(outlier)
```

```python
Q1=df['price'].quantile(0.25)
Q1
```

```python
Q3=df['price'].quantile(0.75)
Q3
```

```python
iqr=Q3-Q1
iqr
```

```python
ul=Q3+1.5*iqr
ul
```

```python
ll=Q1-1.5*iqr
ll
```

```python
df[(df['price']>ul) | (df['price']<ll )]
```

```python
df['price']=np.where((df['price'] > ul),
                                       ul,
                                       np.where((df['price'] < ll),
                                               ll,
                                                df['price'] ))
```

```python
sns.boxplot(df['price'])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OrdinalEncoder
```

```python
transformer=ColumnTransformer(transformers=[
    ('t2',MinMaxScaler(),['price']),
    ('t0',StandardScaler(),['area']),
    ('t1',OneHotEncoder(sparse=False,drop='first'),['mainroad','guestroom','basement','airconditioning']),
    ('t3',OrdinalEncoder(categories=[['unfurnished','semi-furnished','furnished']]),['furnishingstatus'])
    ],remainder='passthrough')
```

```python
df_new=transformer.fit_transform(df)
df_new.shape
```

```python
df_new
```

```python
df_new=pd.DataFrame(df_new,columns=['price','area','mainroad','guestroom','basement','airconditioning','furnishingstatus','bedroom','bathroom','stories','parking'])
```
