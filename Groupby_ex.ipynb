```python
#drop will delete the column that is provided
df1=df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
df1
```

```python
df1.describe()
```

```python
df1.isnull().sum()
```

```python
#fillna(method='ffill') will fill the NaN(none) value with the previous value, ffill(forwerd fill)
df2=df1.fillna(method='ffill')
df2
```

```python
df2.isnull().sum()
```

```python
#groupby return the number of survivied and not survived in the table , where as 0.0-not survived and 0.1-survived
survive=df2.groupby('Survived')['Survived'].count()
survive
```

```python
#groupby pclass , return the no` of passenger in each class
Pclass=df2.groupby('Pclass')['Pclass'].count()
Pclass
```

```python
#groupby sex , return the no` male and female passenger 
sex=df2.groupby('Sex')['Sex'].count()
sex
```

```python
#groupby sex survived , return the no` of male and female survivied 
sex_survive=df2.groupby('Sex')['Survived'].sum()
sex_survive
```

```python
#groupby embarked return the no` of passenger embarked from each place
embark=df2.groupby('Embarked')['Embarked'].count()
embark
```

```python
tab_c_g_s=df2.groupby(['Pclass','Sex','Survived'])['Survived'].count()
tab_c_g_s
```

```python
tab_c_s=df2.groupby('Pclass')['Survived'].sum()
tab_c_s
```

```python
tab_c_s=df2.groupby('Pclass')['Survived'].count()
tab_c_s
```

```python
#this code return the percentage of survived according to each class
tab_c_s_percent=df2.groupby('Pclass')['Survived'].sum()/df2.groupby('Pclass')['Survived'].count()*100
tab_c_s_percent
```

```python
tab_c_g_s=df2.groupby(['Pclass','Sex'])['Survived'].sum()
tab_c_g_s
```
