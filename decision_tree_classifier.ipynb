```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("dataset/iris.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['species']=le.fit_transform(df['species'])
```

```python
df.sample(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,0:4]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion='entropy',splitter='random')
```

```python
dtc.fit(x_train,y_train)
```

```python
pred=dtc.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('accuracy of decision tree',accuracy_score(y_test,pred))
```

```python
print('decision tree confusion matrix')
pd.DataFrame(confusion_matrix(y_test,pred),columns=list(range(0,3)))
```

```python
result=pd.DataFrame()
result['actual label']=y_test
result['decision tree predicition']=pred
```
